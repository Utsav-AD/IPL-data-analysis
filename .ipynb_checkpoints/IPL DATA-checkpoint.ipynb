```python
#loading the required libraries 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#loading ipl dataset
ipl=pd.read_csv('matches.csv')
```

```python
#showing first five record of dataset
ipl.head()
```

```python
#looking at the number of rows and cloumn in the dataset 
ipl.shape
```

```python
#getting the frequency of most man of the match awards
ipl['player_of_match'].value_counts()
```

```python
#getting the top 10 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
#getting the top 5 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:5]
```

```python
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
#making a bar plot for the top 5 players with most man of the match awards
plt.figure(figsize=(10,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()), list(ipl['player_of_match'].value_counts()[0:5]))
plt.show()
```

```python
#getting the frequency of result column
ipl['result'].value_counts()
```

```python
#finiding the number of toss wins each team 
ipl['toss_winner'].value_counts()
```

```python
#finding out the number of matching won when batting first 
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
batting_first.shape
```

```python
#making histogram
plt.figure(figsize=(5,7))
plt.hist(batting_first['win_by_runs'])
plt.title("distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
#finding out the number of wins team have after batting first 
batting_first['winner'].value_counts()
```

```python
#making bar graph of top 5 team winning the most after batting first
plt.figure(figsize=(12,6))
teams=list(batting_first['winner'].value_counts()[0:5].keys())
numberOfWins= list(batting_first['winner'].value_counts()[0:5])

plt.bar(teams, numberOfWins, width=0.2, align= 'center', edgecolor='grey')
plt.ylabel('Number of Wins')
plt.xlabel('Team Names', color='red', weight='600', size='20' )
plt.show()
```

```python
#making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()), autopct='%0.1f%%')
plt.show()
```

```python
#analysing the wins after batting second
ipl.head()

batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
#looking at top 5 data"
batting_second.head()
```

```python
#making histogram for frequency of wins wrt number of wickets
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'], bins=range(1, 12), color='skyblue', edgecolor='black', alpha=0.7, rwidth=0.8)
plt.xlabel("Wickets Remaining")
plt.ylabel("Frequency of Wins")
plt.title("Distribution of Wins by Wickets Remaining")
plt.grid(axis='y', linestyle='--', alpha=1)
plt.show()
```

```python
#finding out the frequency of number of wins when batting second
batting_second['winner'].value_counts()
```

```python
#making bar graph for top 5 teams with most wins
plt.figure(figsize=(12,6))
colors = ['royalblue', 'tomato', 'seagreen', 'goldenrod', 'purple'] 
plt.bar(list(batting_second['winner'].value_counts()[0:5].keys()),list(batting_second['winner'].value_counts()[0:5]) ,color=colors, edgecolor='black', alpha=0.85, width=0.2)
plt.xlabel("Teams", fontsize=12, fontweight='bold')
plt.ylabel("Number of Wins", fontsize=12, fontweight='bold')
plt.title("Top 5 Teams with Most Wins", fontsize=14, fontweight='bold', color='darkred')

plt.xticks(rotation=30, fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#making pie chart of most for distribution of most wins after batting second 
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()), autopct='%0.1f%%')
plt.show
```
